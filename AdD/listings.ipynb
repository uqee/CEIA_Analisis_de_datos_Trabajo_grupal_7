# 1. Exploración y comprensión de los datos

## 1.1. Descargar

```python
import urllib.request
urllib.request.urlretrieve("https://data.insideairbnb.com/argentina/ciudad-aut%C3%B3noma-de-buenos-aires/buenos-aires/2026-01-25/visualisations/listings.csv", "listings.csv")
```

```python
import pandas
dataFrame = pandas.read_csv('./listings.csv')
```

## 1.2. Verificar

```python
print(dataFrame.head())
```

```python
print(dataFrame.tail())
```

## 1.3. Explorar

```python
print("Tamaño")
print(dataFrame.size)
```

```python
print("Tipos de información")
print(dataFrame.dtypes)
```

```python
print("Curiosidades: Datos duplicados")
print(dataFrame.duplicated().sum())
```

```python
print("Desafíos: Valores faltantes")
print(dataFrame.isnull().sum())
```

```python
cols_relevantes = ["latitude", "longitude", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365", "number_of_reviews_ltm"]
print("Estadísticas descriptivas")
print(dataFrame[cols_relevantes].describe())
```

```python
print("Curiosidades: Tipos de alojamiento")
print(dataFrame["room_type"].value_counts())
```

```python
print("Curiosidades: Rango temporal de reseñas")
reviews = pandas.to_datetime(dataFrame["last_review"], errors="coerce")
print("  Primera reseña:", reviews.min())
print("  Última reseña: ", reviews.max())
print("  Rango (días):  ", (reviews.max() - reviews.min()).days)
```

```python
sin_disponibilidad = (dataFrame["availability_365"] == 0).sum()
print("Curiosidades: Listados sin disponibilidad")
print(f"  {sin_disponibilidad} listados ({sin_disponibilidad / len(dataFrame) * 100:.1f}%) tienen availability_365 = 0")
```

```python
print("Desafíos: Precio nulo en todos los registros")
null_pct = dataFrame["price"].isnull().sum() / len(dataFrame) * 100
print(f"  Filas sin precio: {dataFrame['price'].isnull().sum()} ({null_pct:.1f}%)")
```
